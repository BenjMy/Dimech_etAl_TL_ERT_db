# My plot 2 with matplolib

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1, 3, 4])
y = np.array([1, 5, 3])
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.scatter(x, y)
plt.show()                                            
```
